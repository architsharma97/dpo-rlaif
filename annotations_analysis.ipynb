```python
import os
import json

c1_prefs_path = '/ebs/.cache/ubuntu/final_ds_llama7b_1turn512_sft_bs8lr1e-6_df0.1_2023-10-06_17-11-26_207956/epoch-3/comparisons_claude/temp1.0_vs_chatgpt'
c2_prefs_path = '/ebs/.cache/ubuntu/final_ds_llama7b_1turn512_sft_bs8lr1e-6_df0.1_2023-10-06_17-11-26_207956/epoch-3/comparisons_claude_2/temp1.0_vs_chatgpt'
gpt4_prefs_path = '/ebs/.cache/ubuntu/final_ds_llama7b_1turn512_sft_bs8lr1e-6_df0.1_2023-10-06_17-11-26_207956/epoch-3/comparisons_gpt4/temp1.0_vs_chatgpt'
```

```python
c1_prefs = json.load(open(os.path.join(c1_prefs_path, 'annotations.json')))
c2_prefs = json.load(open(os.path.join(c2_prefs_path, 'annotations.json')))
gpt4_prefs = json.load(open(os.path.join(gpt4_prefs_path, 'annotations.json')))
total_prefs = len(c1_prefs)
```

```python
c1_wr, c2_wr, gpt4_wr = 0, 0, 0
agreement = 0
agg_c1c2, agg_c1gpt4, agg_c2gpt4 = 0, 0, 0
for idx in range(total_prefs):
    assert c1_prefs[idx]['output_1'] == c2_prefs[idx]['output_1'] == gpt4_prefs[idx]['output_1']
    assert c1_prefs[idx]['output_2'] == c2_prefs[idx]['output_2'] == gpt4_prefs[idx]['output_2']
    if c1_prefs[idx]['preference'] == 2:
        c1_wr += 1
    if c2_prefs[idx]['preference'] == 2:
        c2_wr += 1
    if gpt4_prefs[idx]['preference'] == 2:
        gpt4_wr += 1
    
    if c1_prefs[idx]['preference'] == c2_prefs[idx]['preference'] == gpt4_prefs[idx]['preference']:
        agreement += 1
    if c1_prefs[idx]['preference'] == c2_prefs[idx]['preference']:
        agg_c1c2 += 1
    if c1_prefs[idx]['preference'] == gpt4_prefs[idx]['preference']:
        agg_c1gpt4 += 1
    if c2_prefs[idx]['preference'] == gpt4_prefs[idx]['preference']:
        agg_c2gpt4 += 1

print(c1_wr / total_prefs, c2_wr / total_prefs, gpt4_wr / total_prefs)
print(agreement / total_prefs)
print(agg_c1c2 / total_prefs, agg_c1gpt4 / total_prefs, agg_c2gpt4 / total_prefs)
```
